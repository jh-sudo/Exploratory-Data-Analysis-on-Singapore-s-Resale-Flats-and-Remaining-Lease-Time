###  5.2.1 North Region

```python
# Map the 'Region' based on the 'town'
df['Region'] = df['town'].map(town_to_region)

# Sort the DataFrame by lease commencement date
df.sort_values('lease_commence_date', inplace=True)

# Set the plotting style
sns.set(style="whitegrid")

# Filter data for the 'North' region
north_data = df[df['Region'] == 'North']

# Create a scatter plot for the 'North' region
plt.figure(figsize=(12, 6))
sns.scatterplot(x='lease_commence_date', y='resale_price', data=north_data, color='red', alpha=0.7, edgecolor='w')
plt.title('North Region: Lease Commencement Date vs Resale Price')
plt.xlabel('Lease Commencement Date')
plt.ylabel('Resale Price')
plt.show()
```

#### Observations

  - The scatter plot illustrates a notable peak in resale prices within the period 1992 to 1995.
  - During this timeframe, the prices reached their highest point before gradually declining.

- **Price Decline (1995-2010):**
  - Following the peak, there is a discernible decline in resale prices from 1995 to 2010.
  - This period represents a gradual tapering off of prices, indicating a potential market correction.

- **Resale Price Recovery (2010 Onward):**
  - After 2010, there is a noticeable reversal in the trend, with resale prices starting to rise again.
  - The recovery suggests a renewed demand or other influencing factors contributing to the upward trajectory.

- **Wave-Like Pattern:**
  - The overall pattern resembles a wave, starting with lower prices, peaking around 1995, tapering off, and then gradually rising again post-2010.
  - This wave-like pattern could be indicative of cyclical market trends or external economic factors influencing the real estate market.

#### Conclusion
The North region's resale market exhibits a dynamic pattern, with distinct phases of rise, decline, and recovery. Further investigation into the external economic factors or policy changes during these periods could provide a deeper understanding of the observed trends.

## 7. Analysis of Resale Prices: Identifying High-Value Regions and Towns

### Overview
In this segment of the analysis, our attention shifts to the exploration of resale prices across different towns and regions within Singapore. The primary goal is to identify the specific towns or regions that consistently exhibit the highest resale prices for residential flats. Through a comprehensive examination of resale data, I aim to pinpoint the areas that stand out in terms of property value, allowing us to draw meaningful conclusions about the factors influencing these high resale prices.

```python
# Define custom colors for each region
region_colors = {'North': 'red', 'West': 'green', 'Central': 'orange', 'North-East': 'purple', 'East': 'blue'}

# Group by region and calculate the average resale price
region_avg_prices = df.groupby('Region')['resale_price'].mean().sort_values(ascending=False)

# Plot the bar graph with custom colors
plt.figure(figsize=(12, 6))
sns.barplot(x=region_avg_prices.index, y=region_avg_prices.values, palette=[region_colors[region] for region in region_avg_prices.index])
plt.title('Average Resale Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Resale Price')
plt.show()
```

## 8. Analyzing the Relationship Between Floor Area, Resale Price, and Lease Commencement Date

### Introduction:
In this analysis, I aim to explore the intricate relationship between the floor area of resale flats, their corresponding resale prices, and the lease commencement dates. Understanding these dynamics can provide insights into how various factors contribute to the resale market trends in Singapore. The floor area serves as a proxy for the size of the flat, while the resale price reflects the monetary value at which these properties are transacted. Additionally, the lease commencement date provides a temporal context, allowing us to examine how the age of a flat influences its resale price. By visualizing these variables, I hope to uncover patterns, trends, and potential correlations that contribute to the nuanced landscape of Singapore's resale flat market.

```python
# Plot the scatter graph
plt.figure(figsize=(12, 8))
scatter_plot = sns.scatterplot(x='floor_area_sqm', y='resale_price', hue='lease_commence_date', data=df, palette='viridis', alpha=0.7, edgecolor='w')
plt.title('Scatter Plot: Floor Area vs Resale Price (Color: Lease Commencement Date)')
plt.xlabel('Floor Area (sqm)')
plt.ylabel('Resale Price')
plt.legend(title='Lease Commencement Date', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

**Insights from the Analysis:**

The scatter plot visualizing the relationship between floor area, resale price, and lease commencement date reveals several intriguing insights. One notable observation is that newer flats with smaller floor areas are often sold at higher prices. This suggests that contemporary buyers might place greater emphasis on factors other than sheer size, such as modern amenities, location, or design features.

Moreover, there are instances where smaller, newer flats command higher prices than larger flats constructed a decade or two ago. This phenomenon could be attributed to various factors, including changing preferences, increased demand for newer developments, or the presence of unique selling points in these modern units.

The diverse patterns in the scatter plot underscore the complexity of the resale market, where multiple factors converge to influence transaction prices. Analyzing these trends provides valuable insights for potential homebuyers, sellers, and policymakers seeking to understand the dynamics of the Singapore resale flat market.


## Conclusion: Insights and Interpretation

The observed trend in the graph suggests a correlation between newer flats having smaller floor areas but potentially higher resale prices compared to larger, older flats. Several factors may contribute to this phenomenon:

1. **Location and Amenities:** Newer flats may be strategically located in areas with improved infrastructure, proximity to schools, public transportation, or other amenities. Buyers may be willing to pay a premium for the convenience and accessibility offered by these modern developments.

2. **Quality of Construction:** Newer flats are often built with modern construction techniques, materials, and design considerations. The perceived quality and aesthetics of these flats could contribute to higher resale prices.

3. **Government Policies:** Changes in government policies or incentives may influence developers to focus on constructing smaller, more efficient flats. These policies can affect the supply-demand dynamics, impacting resale prices.

4. **Market Trends:** Consumer preferences and market trends may shift over time. There might be a growing demand for smaller, more manageable living spaces, influencing developers to build such units.

5. **Supply and Demand Dynamics:** The relationship between supply and demand in the resale market can significantly impact prices. Limited availability of newer flats with smaller floor areas could drive up prices.

6. **Investment Potential:** Buyers may view newer, smaller flats as better long-term investments, expecting appreciation in value over time. This perception can contribute to higher resale prices.

7. **Renovation and Maintenance Costs:** Newer flats might require less immediate renovation or maintenance, which could be attractive to buyers willing to pay a premium for a hassle-free property.

Understanding these potential factors requires a more in-depth analysis of market conditions, buyer behavior, and regional variations. It's essential to consider the holistic context of Singapore's real estate landscape to draw comprehensive conclusions about the observed correlation.
